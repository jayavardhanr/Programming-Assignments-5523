```python
import numpy as np
from math import sqrt,pi,exp
import math

np.random.seed(1)

mean = [1, 0]
cov = [[1, 0], [0, 1]]
p_x_y1 = np.random.multivariate_normal(mean, cov, 50)

mean = [-1, 0]
cov = [[1, -1/2], [-1/2, 1]]
p_x_y_minus1 = np.random.multivariate_normal(mean, cov, 50)

import matplotlib.pyplot as plt
x1 = np.linspace(-4,4,100)
x2 = np.linspace(-4,4,100) 
line = (((x1**2)/3.0)+(14.0*(x1/3.0))+((x2**2)/3.0)+((4.0*x2)/3.0)+((4.0*x1*x2)/3.0)+(1.0/3.0)-math.log(3))

x1_pos, x2_pos = p_x_y1.T
x1_neg, x2_neg = p_x_y_minus1.T

plt.figure(figsize=(10,10))

plt.plot(x1_pos, x2_pos,'b+',x1, line,'g-',x1_neg, x2_neg,'r_')
plt.title('Bayes Decision boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(['50 Test points from p(x,y=1)','Bayes Decision Boundary','50 Test points from p(x,y=-1)'])
plt.show()
```

```python
#Test Error
count_false_negatives=0
count_false_positives=0
count_true_positives=0
count_true_negatives=0

for x1,x2 in p_x_y1:
    if (((x1**2)/3.0)+(14.0*(x1/3.0))+((x2**2)/3.0)+((4.0*x2)/3.0)+((4.0*x1*x2)/3.0)+(1.0/3.0)-math.log(3))<0:
        count_false_negatives+=1
    else:
        count_true_negatives+=1
        
            
for x1,x2 in p_x_y_minus1:
    if (((x1**2)/3.0)+(14.0*(x1/3.0))+((x2**2)/3.0)+((4.0*x2)/3.0)+((4.0*x1*x2)/3.0)+(1.0/3.0)-math.log(3))>=0:
        count_false_positives+=1
    else:
        count_true_positives+=1
        
        
#print('false_negatives:',count_false_negatives)
#print('false_positives:',count_false_positives)
#print('true_positives:',count_true_positives)
#print('true_negatives:',count_true_negatives)

#print('total_test_points:',100)

test_error_bayes=(count_false_negatives+count_false_positives)/(len(p_x_y1)+len(p_x_y_minus1))
print('Test_error for Bayes Rule:',test_error_bayes)
```

```python
import numpy as np
from math import sqrt,pi,exp
import math

np.random.seed(2)

mean = [1, 0]
cov = [[1, 0], [0, 1]]
p_x_y1_LDA = np.random.multivariate_normal(mean, cov, 50)

mean = [-1, 0]
cov = [[1, -1/2], [-1/2, 1]]
p_x_y_minus1_LDA = np.random.multivariate_normal(mean, cov, 50)

mean_pos=p_x_y1_LDA.mean(axis=0)
mean_neg=p_x_y_minus1_LDA.mean(axis=0)
covariance_pos=np.cov(p_x_y1_LDA,rowvar=False)
covariance_neg=np.cov(p_x_y_minus1_LDA,rowvar=False)
c_w=covariance_pos+covariance_neg

#Projection
w=np.dot(np.linalg.inv(c_w),(mean_pos-mean_neg))

import matplotlib.pyplot as plt
x1 = np.linspace(-4,4,100) 
x2 = np.linspace(-4,4,100) 

line_LDA=w[0]*x1+w[1]*x2

x1_pos_LDA, x2_pos_LDA = p_x_y1_LDA.T
x1_neg_LDA, x2_neg_LDA = p_x_y_minus1_LDA.T

plt.figure(figsize=(10,10))

plt.plot(x1_pos_LDA, x2_pos_LDA,'b+',x1,line_LDA,'m-',x1_neg_LDA, x2_neg_LDA,'r_')
plt.title('LDA projection')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(['50 Training points from p(x,y=1) for LDA','LDA projection','50 Training points from p(x,y=-1) for LDA'])
plt.show()
```

```python
import numpy as np
from math import sqrt,pi,exp
import math
import matplotlib.pyplot as plt

#bayes 
np.random.seed(1)

mean = [1, 0]
cov = [[1, 0], [0, 1]]
p_x_y1 = np.random.multivariate_normal(mean, cov, 50)

mean = [-1, 0]
cov = [[1, -1/2], [-1/2, 1]]
p_x_y_minus1 = np.random.multivariate_normal(mean, cov, 50)

x1_pos, x2_pos = p_x_y1.T
x1_neg, x2_neg = p_x_y_minus1.T


#LDA
np.random.seed(2)

mean = [1, 0]
cov = [[1, 0], [0, 1]]
p_x_y1_LDA = np.random.multivariate_normal(mean, cov, 50)

mean = [-1, 0]
cov = [[1, -1/2], [-1/2, 1]]
p_x_y_minus1_LDA = np.random.multivariate_normal(mean, cov, 50)

mean_pos=p_x_y1_LDA.mean(axis=0)
mean_neg=p_x_y_minus1_LDA.mean(axis=0)
covariance_pos=np.cov(p_x_y1_LDA,rowvar=False)
covariance_neg=np.cov(p_x_y_minus1_LDA,rowvar=False)
c_w=covariance_pos+covariance_neg

#Projection
w=np.dot(np.linalg.inv(c_w),(mean_pos-mean_neg))
#print(w)
x1 = np.linspace(-4,4,100) 
x2 = np.linspace(-4,4,100) 

line_LDA=w[0]*x1+w[1]*x2
line = (((x1**2)/3.0)+(14.0*(x1/3.0))+((x2**2)/3.0)+((4.0*x2)/3.0)+((4.0*x1*x2)/3.0)+(1.0/3.0)-math.log(3))

plt.figure(figsize=(10,10))

plt.plot(x1_pos, x2_pos,'b+',x1, line,'g-',x1, line_LDA,'m-',x1_neg, x2_neg,'r_')
plt.title('Bayes Decision boundary vs LDA')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(['random points from p(x,y=1)','Bayes Decision Boundary','LDA Projection','random points from p(x,y=-1)'])
plt.show()
```

```python
#Test Error for LDA
count_false_negatives_LDA=0
count_false_positives_LDA=0
count_true_positives_LDA=0
count_true_negatives_LDA=0

for x1,x2 in p_x_y1:
    if (w[0]*x1+w[1]*x2)<0:
        count_false_negatives_LDA+=1
    else:
        count_true_negatives_LDA+=1
        
            
for x1,x2 in p_x_y_minus1:
    if (w[0]*x1+w[1]*x2)>=0:
        count_false_positives_LDA+=1
    else:
        count_true_positives_LDA+=1
        
        
#print('false_negatives:',count_false_negatives_LDA)
#print('false_positives:',count_false_positives_LDA)
#print('true_positives:',count_true_positives_LDA)
#print('true_negatives:',count_true_negatives_LDA)

#print('total_test_points:',100)

test_error_LDA=(count_false_negatives_LDA+count_false_positives_LDA)/(len(p_x_y1)+len(p_x_y_minus1))
print('test_error for LDA Projection:',test_error_LDA)

Difference_between_LDA_and_bayes=test_error_bayes-test_error_LDA

if Difference_between_LDA_and_bayes>0:
    print('LDA is less than Bayes error by:',abs(Difference_between_LDA_and_bayes))
elif Difference_between_LDA_and_bayes<0:
    print('LDA is greater than Bayes error by:',abs(Difference_between_LDA_and_bayes))
else:
    print('LDA is equal to Bayes error ')
```
