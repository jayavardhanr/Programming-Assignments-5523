# Question 3

```python
import numpy as np
from math import sqrt,pi,exp

#training Data
np.random.seed(1)

x=np.random.randn(1000,1)
ones = np.ones((1000,1))
x_=np.append(ones,x,axis=1)
z=sqrt(0.26)*np.random.randn(1000,1)

#testing Data
x_test=np.random.randn(100,1)
ones = np.ones((100,1))
x_test_=np.append(ones,x_test,axis=1)
z_test=sqrt(0.26)*np.random.randn(100,1)

y=1-(0.5*x)+z
y_test=1-(0.5*x_test)+z_test
w=np.linalg.inv(np.matrix.transpose(x_).dot(x_)).dot(np.matrix.transpose(x_).dot(y))

print('w',w)
print('optimal model:',str(w[0][0])+str(w[1][0])+'x')

#Training Error
y_predict=x_.dot(w)
mse = ((y - y_predict) ** 2).mean()
print("Training Error:",mse)

#Test Error
y_predict_test=x_test_.dot(w)
mse_test = ((y_test - y_predict_test) ** 2).mean()
print("Test Error:",mse_test)

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#f(x)
line = w[0][0] + w[1][0]* x 
plt.plot(x, y,'b+',x, line,'r--',label='f(x)')
plt.title('Estimated Model')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['Training Data','Optimal Model(f(x))'])
plt.show()

#g(x)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
gx=(1.0/sqrt(2*pi*0.26))* np.exp(-np.square(np.add(y,((0.5*x)-1)))/(2*0.26))
ax.scatter(x, y, gx, c='r', marker='o')
ax.view_init(5,azim=135)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('g(x)')
ax.legend(['MSE(g(x))'])
plt.show()
```
