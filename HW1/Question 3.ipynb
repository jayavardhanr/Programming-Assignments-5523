```python
import numpy as np
from math import sqrt,pi,exp
```

```python
#training Data
np.random.seed(1)

x=np.random.randn(1000,1)
ones = np.ones((1000,1))
x_=np.append(ones,x,axis=1)
z=sqrt(0.26)*np.random.randn(1000,1)
```

```python
#testing Data
x_test=np.random.randn(100,1)
ones = np.ones((100,1))
x_test_=np.append(ones,x_test,axis=1)
z_test=sqrt(0.26)*np.random.randn(100,1)
```

```python
y=1-(0.5*x)+z
```

```python
y_test=1-(0.5*x_test)+z_test
```

```python
w=np.linalg.inv(np.matrix.transpose(x_).dot(x_)).dot(np.matrix.transpose(x_).dot(y))
```

```python
w
```

```python
#train Error
```

```python
y_predict=x_.dot(w)
mse = ((y - y_predict) ** 2).mean()
mse
```

```python
#Test Error
```

```python
y_predict_test=x_test_.dot(w)
mse_test = ((y_test - y_predict_test) ** 2).mean()
mse_test
```
