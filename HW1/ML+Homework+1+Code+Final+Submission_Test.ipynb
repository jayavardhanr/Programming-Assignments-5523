# Question 3

```python
import numpy as np
from math import sqrt,pi,exp

#training Data
np.random.seed(1)

x=np.random.randn(1000,1)#with variance 1
ones = np.ones((1000,1))
x_=np.append(ones,x,axis=1)
z=sqrt(0.01)*np.random.randn(1000,1)#with variance 0.01

#testing Data
x_test=np.random.randn(100,1)
ones = np.ones((100,1))
x_test_=np.append(ones,x_test,axis=1)
z_test=sqrt(0.01)*np.random.randn(100,1)

y=1-(0.5*x)+z
y_test=1-(0.5*x_test)+z_test
w=np.linalg.inv(np.matrix.transpose(x_).dot(x_)).dot(np.matrix.transpose(x_).dot(y))

print('w',w)
print('estimated model f(x):',str(w[0][0])+str(w[1][0])+'x')
print('optimal model g(x):',str(1)+str(-0.5)+'x')

#Training Error
y_predict=x_.dot(w)
mse = ((y - y_predict) ** 2).mean()
print("Training Error:",mse)

#Test Error
y_predict_test=x_test_.dot(w)
mse_test = ((y_test - y_predict_test) ** 2).mean()
print("Test Error:",mse_test)

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#f(x)
line = w[0][0] + w[1][0]* x 

#MSE
gx=1 +(-0.5)* x 

plt.figure(figsize=(15,15))
plt.plot(x, y,'b+',x, line,'r--',x, gx,'g--')
plt.title('Estimated Model- f(x)  vs optimal model- g(x) ')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['Training Data','Estimated Model- f(x)','Optimal Model- g(x)'])
plt.show()
```
