```python
from numpy import loadtxt
import numpy as np
import math
import matplotlib.pyplot as plt
import pprint
import pandas as pd
from IPython.display import display, HTML
```

```python
#Loading Data Files
#Features
trainingData = loadtxt("/Users/jayavardhanreddy/Python_Code/ML/HW4/DataFiles/trainingData.txt", delimiter=",",unpack=False,dtype=np.float128)
testData = loadtxt("/Users/jayavardhanreddy/Python_Code/ML/HW4/DataFiles/testData.txt", delimiter=",", unpack=False,dtype=np.float128)
validationData = loadtxt("/Users/jayavardhanreddy/Python_Code/ML/HW4/DataFiles/validationData.txt", delimiter=",",unpack=False,dtype=np.float128)

#Labels
trainingLabels = loadtxt("/Users/jayavardhanreddy/Python_Code/ML/HW4/DataFiles/trainingLabels.txt",unpack=False)
testLabels = loadtxt("/Users/jayavardhanreddy/Python_Code/ML/HW4/DataFiles/testLabels.txt",unpack=False)
validationLabels = loadtxt("/Users/jayavardhanreddy/Python_Code/ML/HW4/DataFiles/validationLabels.txt",unpack=False)
```

```python
#Function to calculate gradient
def calculate_gradient(w,lamda):
    x=trainingData
    y=trainingLabels
    gradient=np.zeros((1,784), dtype=np.float128)
    N=len(y)
    for xi,yi in zip(x,y):
        product=yi*np.dot(w,np.transpose(xi))
        if product<1:
            gradient+=(-1*yi)*(xi)
    gradient/=N
    gradient+=lamda*w
    return gradient

#Main Function for Gradient Descent
def gradient_descent(T,learning_rate,lamda):
    w=np.zeros((1,784), dtype=np.float128)
    for t in range(1,T+1):
        gradient=calculate_gradient(w,lamda)
        w-=(learning_rate*gradient)
    return w

#Function to calculate Risk
#Risk calculation - Remove Regularization
def calculate_risk(w,x,y):
    predictions=np.dot(x,np.transpose(w))
    N=len(y)
    y=y.reshape((N,1))
    values=1-np.multiply(y,predictions)
    risk=np.sum(values.clip(0))/N
    return risk

#Function to calculate Classfication Error
def calculate_error(w,x,y):
    N=len(y)
    y=y.reshape((N,1))
    prediction=np.dot(x,w.T)
    prediction[prediction <= 0] = -1.0
    prediction[prediction > 0] = 1.0
    wrong_classification=len([a for a,b in zip(prediction,y) if a!=b])
    error=wrong_classification*100.0/N
    return error

#Function to calculate Classfication Accuracy
def calculate_accuracy(w,x,y):
    N=len(y)
    y=y.reshape((N,1))
    prediction=np.dot(x,w.T)
    prediction[prediction <= 0] = -1.0
    prediction[prediction > 0] = 1.0
    correct_classification=len([a for a,b in zip(prediction,y) if a==b])
    accuracy=correct_classification*100.0/N
    return accuracy
```

```python
#54  different combinations of  Learning rates and lambda
#Training and Validation Erros for each learning rate, lambda
validation_errors=[]
training_errors=[]
learning_rates=[]
lambdas=[]
T=500
for learning_rate in [0.000001,0.000005,0.00001,0.00005,0.0001,0.005,0.001,0.01,1]:
    for lamda in [0.001,0.01,0.1,0,1,10]:
        print('learning_rate',learning_rate,'lamda',lamda)
        w=gradient_descent(T,learning_rate,lamda)
        learning_rates.append(learning_rate)
        lambdas.append(lamda)
        
        validation_error=calculate_error(w,validationData,validationLabels)
        validation_errors.append(validation_error)
        
        training_error=calculate_error(w,trainingData,trainingLabels)
        training_errors.append(training_error)

#Report Generation
df=pd.DataFrame(columns=['Learning_rate','Lambda','Training_Error','Validation_Error'])
df['Learning_rate']=learning_rates
df['Lambda']=lambdas
df['Training_Error']=training_errors
df['Validation_Error']=validation_errors
HTML(df.to_html())
```

```python
#Learning_rate which yields least validation_error
minimum_validation_error=min(validation_errors)
index_of_minimum=validation_errors.index(minimum_validation_error)
learning_rate,lamda=learning_rates[index_of_minimum],lambdas[index_of_minimum]
print("learning_rate for least Validation Error",learning_rate)
print("lambda for least Validation Error",lamda)
print('Validation Error for learning rate:',learning_rate,'and lambda:',lamda,'is:',minimum_validation_error)
```

```python
#Function used to generate list of empirical risk for iterations 1 to T
def Empiricial_risk_vs_iterations(T,learning_rate,lamda):
    w=np.zeros((1,784), dtype=np.float128)
    list_empirical_risk=[]
    for t in range(1,T+1):
        gradient=calculate_gradient(w,lamda)
        w-=(learning_rate*gradient)
        empirical_risk=calculate_risk(w,trainingData,trainingLabels)
        list_empirical_risk.append(empirical_risk)
    return list_empirical_risk

#list of empirical risk values, for iterations from 1 to 1000
T=500
list_empirical_risk=Empiricial_risk_vs_iterations(T,learning_rate,lamda)
```

```python
#Plot Empirical Risk vs Iterations
plt.figure(figsize=(10,10))
plt.plot(range(1,501), list_empirical_risk, 'r--')
plt.ylabel('Empiricial risk')
plt.xlabel('Iterations')
plt.title(' Empirical Risk vs Iterations')
plt.show()

#Find number of iterations which gives minimum risk 
number_of_iterations=list_empirical_risk.index(min(list_empirical_risk))+1
print("Iteration which produces least empirical risk:",number_of_iterations)

print('\nNote:')
print('Empirical Risk reduces as iterations increase as expected')
```

```python
#Final Model
#Best model using the learning rate and number_of_iterations found
w=gradient_descent(number_of_iterations,learning_rate,lamda)
test_error=calculate_error(w,testData,testLabels)
print("best model: learning rate:",learning_rate,'and lamda:',lamda,' for iterations:',number_of_iterations)
print('Test Error of the best model:',test_error)

print('\nNote:')
print('Test Error:',test_error, 'and Validation Error:',minimum_validation_error,'are very close')
```

```python
#Validation Accuracy vs Test Accuracy
training_accuracy=calculate_accuracy(w,trainingData,trainingLabels)
test_accuracy=calculate_accuracy(w,testData,testLabels)
validation_accuracy=calculate_accuracy(w,validationData,validationLabels)
print('training_accuracy',training_accuracy)
print('validation_accuracy',validation_accuracy)
print('test_accuracy',test_accuracy)


print('\nNote:')
print('Validation Accuracy and Test Accuracy are very close')
```
